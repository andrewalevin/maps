```python
# https://github.com/lukes/ISO-3166-Countries-with-Regional-Codes/blob/master/all/all.csv

import pandas as pd

# Specify the path to your CSV file
csv_file_path = 'all.csv'

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv(csv_file_path)

# Display the DataFrame
print(df)
```

```python
grouped = df.groupby('region')
```

```python


for g in grouped:
    print(g)
    filtered_df = df[df['category'] == 'A']
```

```python
unique_keys = df['region'].unique()
print(unique_keys)
```

```python
text = ''

for gr_n in ['Europe', 'Americas','Asia', 'Africa','Oceania',]:
    gr = grouped.get_group(gr_n)
    for item in gr.iterrows():
        text += f'{list(item)[1][2]}\n'

print(text)
```

```python
file_path = "output.txt"

# Open the file in write mode
with open(file_path, 'w') as file:
    # Write each item in the data list to the file
    for line in text.split('\n'):
        file.write(line + '\n')
```

```python
import json

# Example data list
data = ["item1", "item2", "item3"]

# Specify the path to the JSON file
json_file_path = "output.json"

# Convert the list to JSON format
json_data = json.dumps(text.split('\n'))

# Write the JSON data to a file
with open(json_file_path, 'w') as json_file:
    json_file.write(json_data)
```

```python
# Dictionary mapping country names to ISO 3166 alpha-3 codes
country_codes = {
    'Australia': 'AUS',
    'Austria': 'AUT',
    'Canada': 'CAN',
    'China': 'CHN',
    'Estonia': 'EST',
    'Hong Kong': 'HKG',
    'Hungary': 'HUN',
    'India': 'IND',
    'Israel': 'ISR',
    'Luxembourg': 'LUX',
    'Malta': 'MLT',
    'Mexico': 'MEX',
    'New Zealand': 'NZL',
    'Norway': 'NOR',
    'Pakistan': 'PAK',
    'Portugal': 'PRT',
    'Russia': 'RUS',
    'Serbia': 'SRB',
    'Singapore': 'SGP',
    'Slovak Republic': 'SVK',
    'Slovenia': 'SVN',
    'Sweden': 'SWE'
}

# Original text data
text_data = """Australia 0
Austria 0
Canada 0
China 0
Estonia 0
Hong Kong 0
Hungary 0
India 0
Israel 0
Luxembourg 0
Malta 0
Mexico 0
New Zealand 0
Norway 0
Pakistan 0
Portugal 0
Russia 0
Serbia 0
Singapore 0
Slovak Republic 0
Slovenia 0
Sweden 0"""

# Replace country names with ISO 3166 alpha-3 codes
for country, code in country_codes.items():
    text_data = text_data.replace(country, code)

print(text_data)
```
